## Analyse-Architecture 1:

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
instances = [f"Instance {i+1}" for i in range(11)]

latency_mistral = [49.87, 65, 78, 42.95, 69, 50.69, 64, 47.48, 65, 63, 70]
latency_4omini = [47.04, 43.86, 58.98, 30.72, 50.69, 31.44, 55.32, 23.91, 36.91, 45.68, 48.50]

final_value_mistral = [0.184166, 0.230622, 0.183597, 0.095105, 0.091574, 0.138073, 0.246964, 0.399257, 0.272304, 0.254968, 0.164328]
final_value_4omini = [0.0118, 0.0144, 0.0161, 0.0074, 0.0134, 0.0078, 0.0095, 0.0053, 0.0101, 0.0132, 0.0119]

accuracy_mistral = [0.7, 0.8, 0.78, 0.9, 0.9, 0.82, 0.9, 0.82, 0.7, 0.8, 0.9]
accuracy_4omini = [0.7, 0.7, 0.8, 0.9, 0.85, 0.85, 0.7, 0.8, 0.7, 0.7, 0.75]

# Graph 1: Latency Comparison
plt.figure(figsize=(14, 6))
x = np.arange(len(instances))

plt.bar(x - 0.2, latency_mistral, 0.4, label='Mistral-large-8x22B & GPT-4omini', color='royalblue')
plt.bar(x + 0.2, latency_4omini, 0.4, label='GPT-4omini & GPT-4omini', color='orange')

plt.xlabel('RFP Instances')
plt.ylabel('Latency (Seconds)')
plt.title('Latency Comparison Between Architectures')
plt.xticks(x, instances, rotation=90)
plt.legend()

plt.tight_layout()
plt.show()

# Graph 2: Final Value Comparison
plt.figure(figsize=(14, 6))

plt.bar(x - 0.2, final_value_mistral, 0.4, label='Mistral-large-8x22B & GPT-4omini', color='royalblue')
plt.bar(x + 0.2, final_value_4omini, 0.4, label='GPT-4omini & GPT-4omini', color='orange')

plt.xlabel('RFP Instances')
plt.ylabel('Final Value')
plt.title('Final Value Comparison Between Architectures')
plt.xticks(x, instances, rotation=90)
plt.legend()

plt.tight_layout()
plt.show()

# Graph 3: Accuracy Comparison
plt.figure(figsize=(14, 6))

plt.plot(x, accuracy_mistral, label='Mistral-large-8x22B & GPT-4omini', marker='o', color='royalblue')
plt.plot(x, accuracy_4omini, label='GPT-4omini & GPT-4omini', marker='o', color='orange')

plt.xlabel('RFP Instances')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison Between Architectures')
plt.xticks(x, instances, rotation=90)
plt.legend()

plt.tight_layout()
plt.show()

# Graph 4: Accuracy vs Latency
plt.figure(figsize=(10, 6))
plt.scatter(latency_mistral, accuracy_mistral, color='royalblue', label='Mistral-large-8x22B & GPT-4omini', marker='o')
plt.scatter(latency_4omini, accuracy_4omini, color='orange', label='GPT-4omini & GPT-4omini', marker='o')

plt.xlabel('Latency (Seconds)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Latency')
plt.legend()
plt.tight_layout()
plt.show()

# Graph 5: Final Value vs Accuracy
plt.figure(figsize=(10, 6))
plt.scatter(final_value_mistral, accuracy_mistral, color='royalblue', label='Mistral-large-8x22B & GPT-4omini', marker='o')
plt.scatter(final_value_4omini, accuracy_4omini, color='orange', label='GPT-4omini & GPT-4omini', marker='o')

plt.xlabel('Final Value')
plt.ylabel('Accuracy')
plt.title('Final Value vs Accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```
